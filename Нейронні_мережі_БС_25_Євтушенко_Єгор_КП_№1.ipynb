#### Цикл for

```python
lst = [1, 2, 3, 4, 5]
for i in lst:
    print(f"Елемент списку: {i}")
```

#### Цикл while

```python
i = 0
while i < 5:
    print(f"Цикл while, i = {i}")
    i += 1
```

##Відповіді на питання

1. Основні відмінності Python від інших мов програмування. Відступи в Python: Python — це інтерпретована, високорівнева мова програмування з простим і зрозумілим синтаксисом. Однією з його відмінностей є те, що Python використовує відступи для позначення блоків коду, на відміну від дужок у деяких інших мовах (наприклад, Java, C++). Відступи визначають структуру програм і є обов'язковими.
2. Цикли в Python: різниця між for і while:
- Цикл for використовується для ітерації по послідовності (списки, рядки тощо). Він працює до завершення цієї послідовності.
- Цикл while виконується, поки умова залишається істинною. Використовується, коли кількість ітерацій невідома заздалегідь.
3. Умовні конструкції в Python: Умовні конструкції використовуються для прийняття рішень на основі певних умов. Оператор if перевіряє умову, і якщо вона істинна, виконується певний блок коду. Можна використовувати elif для додаткових умов та else, якщо жодна умова не виконалася.
4. Списки в Python: Список — це впорядкована колекція елементів. Створюється за допомогою квадратних дужок, наприклад: lst = [1, 2, 3]. Елементи можна змінювати, додаючи нові через append() або змінюючи на конкретні індекси: lst[0] = 5.
5. Основні типи даних у Python:
 - int: цілі числа, наприклад, x = 10.
 - float: числа з плаваючою комою, наприклад, y = 3.14.
 - str: рядки, наприклад, z = "Hello".
 - bool: булеві значення, наприклад, True або False.
